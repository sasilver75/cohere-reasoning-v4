# Analyzing Results

```python
# Calculate overall completion rate
total_attempts = len(df)
successful_attempts = len(df[df['completion_verification'] == True])
overall_completion_rate = successful_attempts / total_attempts

print(f"Overall completion rate: {overall_completion_rate:.2%}")
print(f"Total attempts: {total_attempts}")
print(f"Successful attempts: {successful_attempts}")
```

```python
# Calculate recovery rate per problem
problem_recovery = df.groupby('row_id')['completion_verification'].mean()

# Plot problem recovery rates
plt.figure(figsize=(12,6))
plt.bar(problem_recovery.index, problem_recovery.values * 100)
plt.xlabel('Problem ID')
plt.ylabel('Recovery Rate (%)')
plt.title('Recovery Rate by Problem')
plt.ylim(0, 100)
plt.grid(True, alpha=0.3)
plt.show()

print(f"\nProblem recovery rate statistics:")
print(f"Mean: {problem_recovery.mean():.2%}")
print(f"Median: {problem_recovery.median():.2%}")
print(f"Min: {problem_recovery.min():.2%}")
print(f"Max: {problem_recovery.max():.2%}")

# Calculate recovery rate per problem-solution pair
solution_recovery = df.groupby(['row_id', 'solution_idx'])['completion_verification'].mean()

# Plot solution recovery rates
plt.figure(figsize=(15,6))
solution_indices = range(len(solution_recovery))
plt.bar(solution_indices, solution_recovery.values * 100)
plt.xlabel('(row_id, solution_idx)')
plt.ylabel('Recovery Rate (%)')
plt.title('Recovery Rate by Problem-IncorrectSolution Pair')
plt.ylim(0, 100)
plt.grid(True, alpha=0.3)

# Create custom x-tick labels showing (row_id, solution_idx)
tick_labels = [f'({idx[0]},{idx[1]})' for idx in solution_recovery.index]
plt.xticks(solution_indices, tick_labels, rotation=45, ha='right')
plt.tight_layout()
plt.show()

print(f"\nProblem-Solution pair recovery rate statistics:")
print(f"Mean: {solution_recovery.mean():.2%}")
print(f"Median: {solution_recovery.median():.2%}")
print(f"Min: {solution_recovery.min():.2%}")
print(f"Max: {solution_recovery.max():.2%}")
```
